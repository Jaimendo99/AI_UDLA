```python
class Disk:
    def __init__(self, size):
        self.size = size


class Tower:
    def __init__(self):
        self.stack: list[Disk] = []
    
    def popDisk(self):
        disk = self.stack.pop()
        return disk

    def emptyStack(self):
        stack = self.stack.copy()
        self.stack = []
        return stack

    def appendStack(self, stack:list[Disk]):
        self.stack += stack

    def pushDisk(self, disk:Disk):

        if not len(self.stack) : self.stack.append(disk); return 1 

        if self.stack[-1].size < disk.size: return 0
        
        else: self.stack.append(disk); return 1


class Layout:
    def __init__(self, stackSize):
        initTower = Tower()

        for i in range(stackSize,0, -1):
            initTower.pushDisk(
                Disk(i)
            )
        self.towerA = initTower
        self.towerB = Tower()
        self.towerC = Tower()
        self.maxHigh = stackSize+2

    def getTower(self,letter):
        if letter  == "a": tower = self.towerA
        elif letter == "b" : tower = self.towerB
        else: tower = self.towerC
        return tower


    def __str__(self):
        string = ""
        string+=f"A: "
        for i in range(self.maxHigh):        
            try: string+=(str(self.towerA.stack[i].size)+"-")
            except: string+="--"
        string+="\n"

        string+=f"B: "
        for i in range(self.maxHigh):        
            try: string+=(str(self.towerB.stack[i].size)+"-")
            except: string+="--"
        string+="\n"

        string+=f"C: "
        for i in range(self.maxHigh):        
            try: string+=(str(self.towerC.stack[i].size)+"-")
            except: string+="--"
        string+="\n"        
        
        return string
            
```

```python
import copy

def solveGame(lo:Layout, target:str, start:str):
    if target == start: return 

    towersLe = ["a", "b", "c"]

    startTower = lo.getTower(start)
    towersLe.remove(start)

    disksNum = copy.copy(startTower.stack.__len__())

    targetTower = lo.getTower(target)
    towersLe.remove(target)

    freeTower = lo.getTower(towersLe[0])

    c = 0
    print(f"Inter {c}\n{lo}")
    
    freeTower.pushDisk(startTower.popDisk())
    c+=1
    print(f"Inter {c}\n{lo}")
    
    while True:
        
        targetTower.pushDisk(startTower.popDisk())
        c+=1
        print(f"Inter {c}\n{lo}")

        targetTower.appendStack(freeTower.emptyStack())
        c+=1
        print(f"Inter {c}\n{lo}")

        if targetTower.stack.__len__() == disksNum: print(f"RESUELTO EN {c} ITERACIONES"); break

        freeTower.appendStack(targetTower.emptyStack())
        c+=1
        print(f"Inter {c}\n{lo}")
```

```python
lo = Layout(4)
solveGame(lo, "b", "a")
```
